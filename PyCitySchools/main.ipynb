```python
# Dependencies and Setup
import pandas as pd
import numpy as np
```

```python
# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="right", on=["school_name", "school_name"])

#Convert dataset to dataframe
df_school = pd.DataFrame(school_data_complete)

#df_school
#df_school.info()
```

# **District Summary**

```python
#Calculate the total number of schools
school_count = df_school['school_name'].nunique()

#Calculate the total number of students
student_count = df_school['student_name'].count()

#Formatting of Currency
def format(x):
    return "${:,.2f}".format(x)

#Calculate the total budget
group_school_s=df_school.groupby(['school_name'])['budget'].mean()
group_school_df = group_school_s.to_frame()
total_budget = group_school_df['budget'].sum()
total_budget = format(total_budget)

#Calculate the average math score 
avg_math_score = df_school['math_score'].mean()

#Calculate the average reading score
avg_reading_score = df_school['reading_score'].mean()

#Calculate the percentage of students with a passing math score (70 or greater)
filter1 = df_school['math_score']>=70
math_70 = df_school['math_score'].where(filter1).count()
math_passing = (math_70/student_count) *100
math_passing = round(math_passing,6)

#Calculate the percentage of students with a passing reading score (70 or greater)
filter2 = df_school['reading_score']>=70
reading_70 = df_school['reading_score'].where(filter2).count()
reading_passing = (reading_70/student_count) *100
reading_passing = round(reading_passing,6)

#Calculate the percentage of students who passed math and reading (% Overall Passing)
overall_70 = df_school[(df_school.reading_score>=70) & (df_school.math_score>=70)].count()
overall_passing = overall_70["Student ID"]
overall_passing = (overall_passing/student_count) *100
overall_passing = round(overall_passing,6)

#Create a dataframe to hold the above results
school_summary_df = pd.DataFrame({"Total Schools": school_count , 
                                  "Total Students": student_count,
                                  "Total Budget": total_budget,
                                  "Average Math Score": avg_math_score,
                                  "Average Reading Score": avg_reading_score,
                                  "% Passing Math": math_passing,
                                  "% Passing Reading": reading_passing,
                                  "% Overall Passing": overall_passing}, index=[0] )

#Format "Total Students" number output
school_summary_df['Total Students'] = school_summary_df['Total Students'].map('{:,.0f}'.format)
#Print Summary
school_summary_df
```

# *School Summary*

```python
#Find school types
df_school_type = df_school.groupby(['school_name'])['type'].unique()
df_school_type = df_school_type.to_frame()
#df_school_type

#Find school Stats1
df_school_stats1 = df_school.groupby(['school_name'])['size','budget','math_score','reading_score'].mean()
#df_school_stats1

#Merge school type with first set of info (stat1)
frames =[df_school_type, df_school_stats1]
df_district1 = pd.concat(frames , axis =1)
#df_district1
```

```python
#% Passing Reading Score
dist_avg_read_df = df_school[(df_school['reading_score']>=70)]
dist_avg_read_df = dist_avg_read_df.groupby(['school_name']).count() / df_school.groupby(['school_name']).count()
dist_avg_read_df = dist_avg_read_df['reading_score'] *100
dist_avg_read_df = dist_avg_read_df.to_frame()
dist_avg_read_df['% Passing Reading'] = dist_avg_read_df['reading_score']
dist_avg_read_df = dist_avg_read_df.drop(columns=['reading_score'])

#% Passing Math Score
dist_avg_math_df = df_school[(df_school['math_score']>=70)]
dist_avg_math_df = dist_avg_math_df.groupby(['school_name']).count() / df_school.groupby(['school_name']).count()
dist_avg_math_df = dist_avg_math_df['math_score'] *100
dist_avg_math_df = dist_avg_math_df.to_frame()
dist_avg_math_df['% Passing Math'] = dist_avg_math_df['math_score']
dist_avg_math_df = dist_avg_math_df.drop(columns=['math_score'])

#% Overall Passing Score
dist_overall_df = df_school[(df_school['math_score']>=70)]
dist_overall_df = dist_overall_df[(dist_overall_df['reading_score']>=70)]
dist_overall_df = dist_overall_df.groupby(['school_name']).count() / df_school.groupby(['school_name']).count()
dist_overall_df = dist_overall_df['math_score'] *100
dist_overall_df = dist_overall_df.to_frame()
dist_overall_df['% Overall Passing'] = dist_overall_df['math_score']
dist_overall_df = dist_overall_df.drop(columns=['math_score'])

#create second set dataframes to hold per student budget
df_school_avg_budget = df_school
#df_school_avg_budget['Per Student Budget'] = df_school.groupby(['school_name'])['budget'].mean() / df_school.groupby(['school_name'])['size'].mean()

df_school_avg_budget = df_school.groupby(['school_name'])['budget'].mean() / df_school.groupby(['school_name'])['size'].mean()
df_school_avg_budget_stat2 = pd.DataFrame()
df_school_avg_budget_stat2 = pd.DataFrame(df_school_avg_budget,columns = ['Per Student Budget'])
```

```python
#Merge all dataframes together to build District Summary
frames2 =[df_district1, dist_avg_math_df,dist_avg_read_df, dist_overall_df,df_school_avg_budget_stat2]
df_district_summary = pd.concat(frames2 , axis =1)

#Format "Total Students" number output
df_district_summary['School Type'] = df_district_summary['type']
df_district_summary['Total Students'] = df_district_summary['size'].map('{:,.0f}'.format)
df_district_summary['Total School Budget'] = df_district_summary['budget'].map('${:,.2f}'.format)
df_district_summary['Per Student Budget'] = df_district_summary['Per Student Budget'].map('${:,.2f}'.format)
df_district_summary['Average Math Score'] = df_district_summary['math_score'].map('{:,.6f}'.format)
df_district_summary['Average Reading Score'] = df_district_summary['reading_score'].map('{:,.6f}'.format)

#Ran once to remove additional info
#df_district_summary_v = df_district_summary.drop(columns=['math_score','reading_score','budget','size',0,"type"])

df_district_summary = pd.DataFrame(df_district_summary, columns=["School Type", "Total Students" , "Total School Budget", "Per Student Budget"
                                                                 ,"Average Math Score","Average Reading Score", "% Passing Math","% Passing Reading","% Overall Passing"] )

df_district_summary
```

# *Math Scores by Grade*

```python
#Calculate 9th grade math score
nine_grade_df = df_school[df_school['grade']=='9th']
nine_grade_df = nine_grade_df.groupby(['school_name']).agg({'math_score':['mean']})
nine_grade_df["9th"] = nine_grade_df["math_score"]
nine_grade_df = nine_grade_df.drop(columns=['math_score'])

#Calculate 10th grade math score
ten_grade_df = df_school[df_school['grade']=='10th']
ten_grade_df = ten_grade_df.groupby(['school_name']).agg({'math_score':['mean']})
ten_grade_df["10th"] = ten_grade_df["math_score"]
ten_grade_df = ten_grade_df.drop(columns=['math_score'])

#Calculate 11th grade math score
eleven_grade_df = df_school[df_school['grade']=='11th']
eleven_grade_df = eleven_grade_df.groupby(['school_name']).agg({'math_score':['mean']})
eleven_grade_df["11th"] = eleven_grade_df["math_score"]
eleven_grade_df = eleven_grade_df.drop(columns=['math_score'])

#Calculate 12th grade math score
twelve_grade_df = df_school[df_school['grade']=='12th']
twelve_grade_df = twelve_grade_df.groupby(['school_name']).agg({'math_score':['mean']})
twelve_grade_df["12th"] = twelve_grade_df["math_score"]
twelve_grade_df = twelve_grade_df.drop(columns=['math_score'])

frames3 =[nine_grade_df,ten_grade_df,eleven_grade_df,twelve_grade_df]
df_math_scores = pd.concat(frames3 , axis =1)
df_math_scores
```

```python
#Create new dataframe to calculate values in bins
df_school_spending_bins = pd.DataFrame()
df_school_spending['Spending Ranges (Per Student)'] = pd.cut(df_school_spending['Per Student Budget'], bin1, labels=group_name1, include_lowest=True)
df_school_spending_bins= df_school_spending

#df_school_spending_bins
```
